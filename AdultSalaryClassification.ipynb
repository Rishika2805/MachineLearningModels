14.native_country â€“ Country of origin (e.g., United States, India, Mexico).

<h3>ðŸ“Œ How to Handle Categorical Features (like native-country)</h3>

<b>Check distribution â†’ See if one category dominates.

Group rare categories â†’ Replace infrequent values with "Other".

Encode â†’ Use One-Hot Encoding (trees) or Label/Ordinal Encoding (linear models).

Drop old column â†’ After encoding, keep only new encoded columns.

Test usefulness â†’ If the feature doesnâ€™t improve accuracy, drop it completely.</b>

```python
cat_col = [col for col in df_encoded.columns if col not in num_col and col != 'income']
cat_col
```

```python
# chi Square test for categorical column
from scipy.stats import chi2_contingency

alpha = 0.05
chi_result = {}

# apply chi square test
for col in cat_col:
    contingency_table = pd.crosstab(df_encoded[col], df_encoded['income'])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    decision = 'Reject Null Hypothesis(Keep Feature)' if p < alpha else 'Accept Null(Drop Feature)'
    chi_result[col] = {'chi2': chi2, 'p_value': p, 'decision': decision}


chi_result_df = pd.DataFrame(chi_result).T
chi_result_df = chi_result_df.sort_values(by='p_value')
chi_result_df
```

```python
chi_col = chi_result_df[chi_result_df['decision'] == 'Reject Null Hypothesis(Keep Feature)'].index.tolist()
chi_col
```

# PipeLine

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.base import BaseEstimator, TransformerMixin
```

```python
class categorize_edu(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    def categorize_education(num):
        if num <= 8:
            return "School Level"
        elif num <= 10:
            return "High School"
        elif num <= 12:
            return "Associate Degree"
        elif num == 13:
            return "Bachelors"
        else:
            return "Postgraduate"

    def transform(self, X):
        X = X.copy()
        X["education_category"] = X["education_num"].apply(categorize_education)
        return X

class hour_cat(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        X = X.copy()
        bins = [0, 19, 34, 40, 60, 99]
        labels = ["Very Part-Time", "Part-Time", "Full-Time", "Over-Time", "Extreme Work"]
        X["hours_category"] = pd.cut(X["hours_per_week"], bins=bins, labels=labels, include_lowest=True)
        return X

class capital_gain(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        X = X.copy()
        gain_bins = [-1, 0, 5000, 10000, df["capital_gain"].max()]
        gain_labels = ["No Gain", "Low Gain", "Medium Gain", "High Gain"]
        X["capital_gain_category"] = pd.cut(X["capital_gain"], bins=gain_bins, labels=gain_labels)
        return X

class capital_loss(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        X = X.copy()
        loss_bins = [-1, 0, 1000, 2000, df["capital_loss"].max()]
        loss_labels = ["No Loss", "Low Loss", "Medium Loss", "High Loss"]
        X["capital_loss_category"] = pd.cut(X["capital_loss"], bins=loss_bins, labels=loss_labels)
        return X

class region(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        X = X.copy()
        X["native_country"] = X["native_country"].replace(
        X["native_country"].value_counts()[X["native_country"].value_counts() < 500].index,
            "Other")
        return X
```

```python
class FeatureSelector(BaseEstimator, TransformerMixin):
    def __init__(self, feature_names):
        self.feature_names = feature_names

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        # If X is a DataFrame
        try:
            return X[self.feature_names]
        # If X is a NumPy array after OneHotEncoder
        except:
            import pandas as pd
            return pd.DataFrame(X).iloc[:, self.feature_names]
```

```python
class ReplaceQuestionMark(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X.replace("?", np.nan)
    def set_output(self, *, transform=None):
        return self
```

```python
def makePipeline(num_att,cat_att):
    numeric_transformer = Pipeline(steps=[
        ("imputer", SimpleImputer(strategy="mean")),
        ('scaler', StandardScaler())
    ])

    categorical_transformer = Pipeline(steps=[
        ("replace_q", ReplaceQuestionMark()),  # NEW STEP
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder(handle_unknown="ignore", sparse_output=False))
    ])

    preprocessor = ColumnTransformer(
        transformers=[
            ('num', numeric_transformer, num_att),
            ('cat', categorical_transformer, cat_att)
        ])
    preprocessor.set_output(transform="pandas")

    full_pipeline = Pipeline([
        ('region', region()),
        ('Educcation_cat', categorize_edu()),
        ('hour_cat', hour_cat()),
        ('capital_gain', capital_gain()),
        ('capital_loss', capital_loss()),
        ('preprocessor', preprocessor),
        ('model', XGBClassifier(use_label_encoder=False, eval_metric='logloss',
                                n_estimators=200, max_depth=4, learning_rate=0.1))
    ])
    return full_pipeline


```
